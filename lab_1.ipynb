LAB_1  : Data Visualization using a Health Dataset Requirement


```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)
df = load_dataset("Health_heart_experimental.csv")
df.head(5)
```

Line Chart

```python
# Group by age and take mean SysBP
df_line = df.groupby('age')['SysBP'].mean().reset_index()

plt.figure()
plt.plot(df_line['age'], df_line['SysBP'], color='green')

plt.title("Average Systolic Blood Pressure vs Age")
plt.xlabel("Age")
plt.ylabel("Average Systolic Blood Pressure")
plt.show()
```

The line graph shows the relationship between age and average 
systolic blood pressure. Because age changes continuously, a line 
graph is suitable for showing overall patterns. The graph shows a 
clear upward trend, indicating that systolic blood pressure tends to 
rise as people get older.

Bar plot

```python
counts = df['indication'].value_counts()

plt.bar(counts.index.astype(str), counts.values)
plt.title("Heart Disease Indication Count")
plt.xlabel("Indication (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

The bar chart shows the number of people who have heart diseases and who have not heart diseases.Bar charts help compare numbers between groups. This chart makes it easy to see if the two groups are similiar in size or if one group larger the other. This chart help for only some type of data shown not more types of data.

PIE CHART

```python
df['sex'].value_counts().plot(kind='pie', autopct='%0.2f')
plt.title("Sex Distribution")
plt.ylabel("")
plt.show()
```

The pie chart shows how the participants are divided by gender. This chart show the  proportion of the male and female. This type of chart shows a quick visual comparison, making it easy to see if one gender has a larger comparison than the other.This tyoe of charts helps to comparison percentage in any type of comparison.

Box plot

```python
sns.boxplot(df['age'])
plt.title("Box Plot of Age")
plt.show()
```

The box plot provides a summary of the age distribution using the five_number aummary: minimum value,first quartile,median,third quartile,and maximum value . It highlights the median asa a measure of central tendency ,while the length of the box represents the interquartile range , showing how the data is spread .The plot helps in identifying any potentials outlier in the dataset.


HISTOGRAM

```python
plt.hist(df['age'])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

The histograms show how ages are spend in the data . The X-axis shows age groups , and the y-axis showa how many people fall into each group .This chart helps us see the overall pattern of ages,such as whether most people are young,old,or spread evenly.

SCATTER PLOT

```python
# Aggregate data
df_scatter = df.groupby('age')['SysBP'].mean().reset_index()

plt.figure()
plt.scatter(
    df_scatter['age'],
    df_scatter['SysBP'],
    s=50
)

plt.title("Average Systolic Blood Pressure vs Age")
plt.xlabel("Age")
plt.ylabel("Average Systolic Blood Pressure")
plt.show()
```

This scatter plot show how age is related to average systolic blood pressure. Each dot represents an age and it's average blood pressure . Scatter plots help show patterns in data .The chart shows that blood pressure usually increases asa age increases.
